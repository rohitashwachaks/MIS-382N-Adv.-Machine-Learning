```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("californiaDDSDataV2.csv")
data.head()
```

```python
data["Age Cohort"].value_counts()
```

```python
data["Ethnicity"].value_counts()
```

```python
data = data[data["Ethnicity"].apply(lambda x: x in ["White not Hispanic","Hispanic"])]#,"Asian","Black"])]
data["Ethnicity"].value_counts()
```

```python
sns.boxplot(data=data, x='Expenditures', y='Ethnicity')
plt.title("Expenditure by Ethnicity")
plt.savefig("./outputs/expenditure-by-ethnicity", transparent = False)
```

```python
def Plot_box(data, age):
    sns.boxplot(data = data[data["Age Cohort"] == age], x= 'Expenditures', y= 'Ethnicity')
    plt.title('Expenditure Comparison for Age Group:'+age)
    plt.savefig('./outputs/expenditure-'+age.replace(" ",""), transparent = False)
    plt.show()
    return
```

```python
for age in data["Age Cohort"].unique():
    Plot_box(data,age)
    # plt.show()
```
